```python
import pandas as pd

df = pd.read_csv('../data/pairs.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
import tifffile


def plot_channels(row):
    # 5 subplots (one per channel), side by side
    fig, axes = plt.subplots(nrows=1, ncols=5, figsize=(15, 4))
    fig.suptitle(f"BROAD_ID: {row['BROAD_ID']}", fontsize=16)

    # Channels in the order we'd like to display
    channels = ["channelDNA", "channelER", "channelRNA", "channelAGP", "channelMito"]
    titles = ["DNA", "ER", "RNA", "AGP", "Mito"]

    for ax, ch_name, title in zip(axes, channels, titles):
        path = row[ch_name]
        img = tifffile.imread(path)

        ax.imshow(img, cmap="gray")
        ax.set_title(title)
        ax.axis("off")

    plt.tight_layout()
    plt.show()
```

```python
plot_channels(df.iloc[0])
plot_channels(df.iloc[1])
plot_channels(df.iloc[2])
```

```python
def plot_by_broad_id(broad_id, site=1, df=df):
    """Plot all channels for a given BROAD_ID and optional site number."""
    # Find all rows for this BROAD_ID
    rows = df[df['BROAD_ID'] == broad_id]

    if len(rows) == 0:
        print(f"No images found for BROAD_ID: {broad_id}")
        return

    # Get unique plate/well combinations
    instances = rows[['plate_id', 'well']].drop_duplicates()
    print(f"Found {len(instances)} instances of {broad_id} across {len(instances['plate_id'].unique())} plates")

    # Plot each instance
    for _, instance in instances.iterrows():
        row = rows[
            (rows['plate_id'] == instance['plate_id']) &
            (rows['well'] == instance['well']) &
            (rows['site'] == site)
        ].iloc[0]

        plot_channels(row)
        plt.show()

def find_compounds(name_pattern):
    """Find compounds where name matches the given pattern (case-insensitive)."""
    matches = df[df['CPD_NAME'].str.contains(name_pattern, case=False, na=False)]
    unique_compounds = matches[['BROAD_ID', 'CPD_NAME', 'CPD_NAME_TYPE']].drop_duplicates()
    return unique_compounds
```
